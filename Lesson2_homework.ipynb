### Урок 2. Дискретные случайные величины. Закон распределения вероятностей.  Биномиальный закон распределения. Распределение Пуассона

```python
import numpy as np
from math import factorial
from IPython.display import Image

# Перестановки
def permutations(n):
    return int(factorial(n))

# Размещения
def arrangements(n, k):
    return int(factorial(n) / factorial(n - k))

# Сочетания
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))

# Биномиальный закон. Формула Бернулли
def binomial(n, k , p):
    return (combinations(n, k)) * (p ** k) * ((1 - p) ** (n - k))

# Формула Пуассона
def poisson_proba(k, lambda_):
    return (lambda_ ** k) * (np.exp(-lambda_)) / np.math.factorial(k)
```

#### Задание 2
Вероятность попадания в цель при одном выстреле равна 0.01. Какова вероятность того, что число попаданий при 200 выстрелах будет не менее 5 и не более 10?

Решение: <br>
Пусть случайная величина X – число попаданий в цель. Так как вероятность p = 0.01 очень мала, а число выстрелов (опытов) достаточно велико, то искомую вероятность будем находить, используя формулу Пуассона. Так как события несовместные, искомая вероятность может быть найдена по формуле сложения вероятностей.

```python
lambda_ = 200 * 0.01
lambda_
```

```python
sum(poisson_proba(k = i, lambda_ = lambda_) for i in range(5, 11))
```

#### Задание 3
В условиях предыдущей задачи сколько выстрелов нужно сделать чтобы быть уверенным с вероятностью 0.9, что хотя бы 10 раз будет совершено попадание?

_Подсказка_. Как посчитать `P(X >= k)` для величины `X`, имеющей распределение Пуассона?

Решение: <br>
P(X>=k) для велечины X равна 1 - P(X< k), следовательно количество промахов при 10 попаданиях будет n - 10. <br>
Значит, наша вероятность равна 0.9 = 1 - p в степени (n - 10). Найдём n путём подбора.

```python
p = 0.1

for i in range(1, 1500):
    n = i
    q = 0.99 ** n
    if 0 < (p - q) < 0.001:
        break
        
print(f'Количество выстрелов равно: {n + 10}')
```
