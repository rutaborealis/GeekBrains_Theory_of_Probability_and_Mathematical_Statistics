### Урок 1. Случайные события. Условная вероятность. Формула Байеса. Независимые испытания

```python
import numpy as np
from math import factorial
from IPython.display import Image

# Перестановки
def permutations(n):
    return int(factorial(n))

# Размещения
def arrangements(n, k):
    return int(factorial(n) / factorial(n - k))

# Сочетания
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

#### Задание 2
Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

Решение: <br>
Необходимо сложить произведения сочетаний. Сначала рассмотрим ситуацию, где выбран 1 туз из 4 карт, остальные 3 карты выбраны из 48 (тузов не рассматриваем), затем - выбрано 2 туза из 4 карт, остальные 2 из 48, затем - выбрано 3 туза из 4 карт, остальная 1  из 48, затем выбрано 4 туза из 4 карт, 0 из 48. Это будут произведения сочетаний.

```python
result = (combinations(4, 1) * combinations(48, 3)) + (combinations(4, 2) * combinations(48, 2)) + (combinations(4, 3) * combinations(48, 1)) + (combinations(4, 4) * combinations(48, 0))
result
```

- Вероятность, что число на кубике больше 3 равна вероятности, что число чётное = 1/2.
- Если известно, что число чётное, то вероятность, что оно больше 3 = 2/3.
- Если известно, что число больше 3, то вероятность, что оно четное = 2/3.
