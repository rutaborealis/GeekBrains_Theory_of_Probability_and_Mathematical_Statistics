#### Задание 1
Даны значения величины заработной платы заемщиков банка (salary) и значения их поведенческого кредитного скоринга (scoring): <br>
salary = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110] <br>
scoring = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832] <br>

1. Найдите ковариацию этих двух величин: сначала без использования специальных функций, а затем с помощью функции numpy.cov. Полученные значения должны быть равны.

2. Найдите коэффициент корреляции Пирсона: с помощью ковариации и средних квадратических отклонений двух признаков, а затем с помощью специальных функций из библиотек numpy и pandas.

Решение: <br>

```python
salary = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
scoring = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])

df_salary_scoring = pd.DataFrame({'Salary': salary, 'Scoring': scoring}, columns=['Salary', 'Scoring'])
df_salary_scoring
```

1) Найдите ковариацию этих двух величин: сначала без использования специальных функций, а затем с помощью функции numpy.cov. Полученные значения должны быть равны.

```python
salary_mean = df_salary_scoring['Salary'].mean()
scoring_mean = df_salary_scoring['Scoring'].mean()

covariance = ((df_salary_scoring['Salary'] - salary_mean) * (df_salary_scoring['Scoring'] - scoring_mean)).sum() / (df_salary_scoring.shape[0] - 1)
covariance
```

```python
np.cov(df_salary_scoring['Salary'], df_salary_scoring['Scoring'])
```

2. Найдите коэффициент корреляции Пирсона: с помощью ковариации и средних квадратических отклонений двух признаков, а затем с помощью специальных функций из библиотек numpy и pandas.

```python
pearson_correlation = covariance / (df_salary_scoring['Salary'].std() * df_salary_scoring['Scoring'].std(ddof = 1))
pearson_correlation
```

```python
df_salary_scoring[['Salary', 'Scoring']].corr()
```

#### Задание 2
Измерены значения IQ выборки студентов, обучающихся в местных технических вузах: <br>
131, 125, 115, 122, 131, 115, 107, 99, 125, 111. <br>
<br>
Известно, что в генеральной совокупности IQ распределен нормально. <br>
Найдите доверительный интервал для математического ожидания с надежностью 0.95.

Решение: <br>

```python
students_iq = np.array([131, 125, 115, 122, 131, 115, 107, 99, 125, 111])

n = students_iq.shape[0]
mean = students_iq.mean()
std = students_iq.std(ddof=1)
p = 0.95
alpha = 1 - p

t1 = stats.t.ppf(alpha / 2, df=n - 1)
t2 = stats.t.ppf(1 - alpha / 2, df=n - 1)

print(f'Доверительный интервал:\n{(mean + t1 * std / np.sqrt(n), mean + t2 * std / np.sqrt(n))}')
```
