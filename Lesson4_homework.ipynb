Используем формулу для нахождения вероятности попадания нормальной случайной величины в интервал: <br>
$P(\alpha < X < \beta) = \Phi\left(\frac{\beta - a}{\sigma}\right) - \Phi\left(\frac{\alpha - a}{\sigma}\right),$
где $\Phi(x)$ - функция Лапласа (значения берутся из таблицы).

$P(182 < X < \infty) = 1 - \Phi\left(\frac{182 - 174}{8}\right)$. <br>

$P(182 < X < \infty) = 1 - 0.3413$, <br>
$P(182 < X < \infty) = 0.6587$.

$P(190 < X < \infty) = 1 - \Phi\left(\frac{190 - 174}{8}\right)$ <br>
$P(X = \infty) = 1, P(190 < X < \infty) = 1 - \Phi(2)$, <br> 
$P(190 < X < \infty) = 1 - 0.4772$, <br>
$P(190 < X < \infty) = 0.5228$. <br>

$P(166 < X < 190) = \Phi\left(\frac{190 - 174}{8}\right) - \Phi\left(\frac{166 - 174}{8}\right)$. <br>
$P(166 < X < 190) = \Phi(2) + \Phi(1)$, <br>

Их средняя масса равна 1.06 кг. <br>

а) среднее квадратическое отклонение массы коробки, <br>

2. Найдите такое значение scale, для которого значение cdf(x=1, loc=1.06, scale=scale) близко к 0.05. Точности 0.0001 будет достаточно.

Решение: <br>
Пусть $X$ - нормально распределенная случайная величина, равная массе коробки с конфетами. <br>
Параметры $a = 1,06$ (математическое ожидание), ${\sigma}$ (среднее квадратичное отклонение). <br>
Используем формулу для нахождения вероятности попадания нормальной случайной величины в интервал: <br>
$P(\alpha < X < \beta) = \Phi\left(\frac{\beta - a}{\sigma}\right) - \Phi\left(\frac{\alpha - a}{\sigma}\right)$, <br>
где $\Phi(x)$ - функция Лапласа (значения берутся из таблицы).

Используем условие: 5% коробок имеют массу, меньшую 1 кг, то есть:

$P(-\infty < X < 166) = \Phi\left(\frac{1 - 1.06}{\sigma}\right) - \Phi\left(\frac{-\infty - 1.06}{\sigma}\right) = 0.05$. <br>
$\Phi\left(\frac{-0.06}{\sigma}\right) - \Phi\left({-\infty}\right) = 0.05$, <br>
$-\Phi\left(\frac{0.06}{\sigma}\right) + 0.5 = 0.05$, <br>
$\Phi\left(\frac{0.06}{\sigma}\right) = 0.45$, <br>
$\frac{0.06}{\sigma} = 1.645$, <br>
${\sigma} = 0.365$ - среднеквадратическое отклонение массы коробки.

Теперь вычислим, каков процент коробок, масса которых превышает 1.1 кг. 

$P(1.1 < X < \infty) = \Phi\left(\frac{\infty - 1.06}{0.365}\right) - \Phi\left(\frac{1.1 - 1.06}{0.365}\right)$. <br>
$\Phi\left({\infty}\right) - \Phi\left({0.109}\right) = 0.05$, <br>
